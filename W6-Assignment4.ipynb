Q3. What is App routing in Flask? Why do we use app routes?

Ans :
    
App routing in Flask describes the procedure of specifying URL endpoints and connecting them to particular functions or views in your application. Routes are the name for these URL endpoints. The Flask application uses the routing mechanism to decide which function or view should handle a user's request to a specific URL.

App routes are used in Flask for several reasons:
1. Request management : You can specify different URL endpoints for various components of your application by using app routes. Each route has a particular function or view that processes the request and produces the necessary output. As a result, you can arrange and structure the logic of your application according to the various routes and their associated functionalities.

2. URL mapping : App routes offer a way to link particular URLs to corresponding Flask application views or functions. You could, for instance, create a route that maps the URL "/home" to a method that displays your application's home page. This facilitates navigation and access to various areas of your application through sensible and simple URLs.

3. RESTful APIs : App routes are essential for specifying the various endpoints for various resources and the corresponding HTTP methods (GET, POST, PUT, DELETE) that can be used on those endpoints when creating RESTful APIs with Flask. By sending requests to the appropriate routes, this enables clients to communicate with the API and carry out CRUD (Create, Read, Update, Delete) operations on the resources.

4. Parameterized routes : In Flask, app routes can also have dynamic URL parameters or segments. The associated function or view can accept these parameters as arguments by extracting them from the URL. This gives you the ability to design flexible routes that can accommodate various inputs or circumstances. For instance, you could create a route called "/users/username>" that accepts any username as an argument and sends it to the appropriate function.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

Ans :

The url_for() function in Flask is used to construct URLs. Based on a given view function's name and any supplied arguments, it generates a URL for it. When creating a URL dynamically, particularly in templates or when redirecting to a specific endpoint, this function comes in handy.


```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Welcome to the homepage!'

@app.route('/user/<username>')
def profile(username):
    return f'Hello, {username}!'

@app.route('/articles')
def articles():
    return 'List of articles'

with app.test_request_context():
  
    print(url_for('index'))
    print(url_for('profile', username='dev'))
    print(url_for('articles'))
```
