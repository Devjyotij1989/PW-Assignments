Q1. Explain GET and POST methods.

Ans :

POST is used to send data to the server for processing, whereas GET is used to retrieve data from the server. Information is requested using GET, while it is submitted using POST.

Q2. Why is request used in Flask?

Ans :

Data sent by the client is handled and accessed by Flask's request object. It enables you to retrieve request headers, query parameters, and form data. It can also be used to handle file uploads and identify the client's HTTP method. In essence, it aids you in interacting with data that a Flask application sends from the client to the server.

Q3. Why is redirect() used in Flask?

Ans :

To direct the user's web browser to a different website, use Flask's'redirect()' function. It is helpful for tasks like directing users to new pages after they submit forms, enforcing specific URL patterns, managing route changes, and enhancing user experience by rerouting users to pertinent pages.

Q4. What are templates in Flask? Why is the render_template() function used?

Ans :

In Flask, templates are HTML documents that contain placeholders for dynamic content. They keep the design and the code separate. These templates are displayed using the'render_template()' function, which also generates the final HTML by combining them with data. Using templates makes it simple to reuse common elements, create dynamic web pages, and keep the code organised.

Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.

```python
from flask import Flask, request

app = Flask(__name__)

# Sample data
users = [
    {"id": 1, "name": "Dev"},
    {"id": 2, "name": "Raj"},
    {"id": 3, "name": "Vik"}
]

# Endpoint to get all users
@app.route('/users', methods=['GET'])
def get_users():
    return users

# Endpoint to get a specific user by ID
@app.route('/users/<int:user_id>', methods=['GET'])
def get_user(user_id):
    user = next((user for user in users if user['id'] == user_id), None)
    if user:
        return user
    else:
        return ({'message': 'User not found'}), 404

# Endpoint to create a new user
@app.route('/users', methods=['POST'])
def create_user():
    data = request.get_json()
    new_user = {'id': len(users) + 1, 'name': data['name']}
    users.append(new_user)
    return new_user

# Running the Flask application
if __name__ == '__main__':
    app.run(host="0.0.0.0")
```
